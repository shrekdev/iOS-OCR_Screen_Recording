```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
conv_to_seconds = lambda td: td.total_seconds()
```

```python
df = pd.read_csv("sample/video.csv")
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values(by=["timestamp"])
first_ts = df["timestamp"][0]
df["duration"] = (df["timestamp"] - df["timestamp"][0]).apply(lambda td: td.total_seconds())
```

```python
fig, axes = plt.subplots()
axes.plot(df["duration"], df["sinr0"], 'g-')
axes.plot(df["duration"], df["sinr1"], 'b-')
axes.set_title("LTE SINR [dB]")
axes.set_xlabel("Elapsed time [s]")
axes.legend()
```
